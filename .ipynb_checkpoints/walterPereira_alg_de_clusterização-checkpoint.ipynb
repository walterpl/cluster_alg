# Clusterizao

Conforme solicitado pelo trabalho, a clusterizao ser feita por Kmean com nmero de clusters setado para 3.

```python
stand_esc = StandardScaler()
column_list = ['child_mort', 'exports', 'health', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp']
stand_esc.fit(country_data_df[column_list])
```

```python
stand_country_data = stand_esc.transform(country_data_df[column_list])
stand_country_data_df = pd.DataFrame(stand_country_data,columns=column_list)
stand_country_data_df.describe().T
```

###### processo de clusterizao

```python
kmean= KMeans(n_clusters=3,random_state=10)
data_fited = kmean.fit(stand_country_data_df)
dist = data_fited.fit_transform(stand_country_data_df)
country_data_df['cluster'] = data_fited.labels_
country_data_df['cluster_dist'] = [dist[x,country_data_df.loc[x,'cluster']] for x in country_data_df.index ]
group_by_cluster = country_data_df.groupby(by='cluster')
```

#### Distribuio por cluster

```python
group_by_cluster.mean().T
```

```python
group_by_cluster.std().T
```

```python
group_by_cluster.min().T
```

```python
group_by_cluster.max().T
```

```python
group_by_cluster.quantile([.25,.5,.75]).T
```

```python
_ = sns.pairplot(country_data_df.loc[:,['child_mort','life_expec','income','cluster','gdpp','total_fer']],hue='cluster', corner=True)
```

O resultado apresenta uma clusterizao soft com rudos pontuais. Em todos os casos os outliers foram agredados ao cluster mais prximo devido a distncia euclidiana, mas as zonas de fronteira entre estes no possuim resoluo clara.
Pela segregao, o algoritmo dividiu os pases entre em desenvolvimento, sub-desenvolvidos e desenvolvidos.

Abaixo esto os pases mais prximos em represento do cluster, os mesmos foram escolhidos baseados na menor distncia euclidiana ao centro do cluster.

## Cluster Hierquico

```python
plt.figure(figsize=(15, 5))
plt.grid(False)
dendrogram = sch.dendrogram(sch.linkage(stand_country_data_df, method='ward'), labels=country_data_df.country.values)
plt.title('Dendrogram')
plt.ylabel('Euclidean Distance')
```

O agrupamento permaneceu em 3, inclusive os pases representantes de cada cluster do exemplo anterior permaneceram cada um dentro de seus respectivo grupo. Entretanto com o dendograme  possvel ter uma observao mais ampla do cenrio global, podendo destacar:
- O maior grupo so dos pases em desenvolvimento, seguido pelos subdesenvolvidos
- Considerando que os dados so um retrato histrico,  possivel verificar pases que esto a margem de tornarem-se subdesenvolvidos e pases que esto prximos  alcanarem padres mnimos dos desenvolvidos, caso dados histricos estivesem disponveis, seria possvel prever quais moveriam-se para qual grupo.
- Os outliers repousam nos grupos extremos, alguns pases destacam-se positiva ou negativamente, em suas features a ponto de serem assimilados a partir da 5 ou at mesmo da ultima layer, logo pases em desenvolvimento tendem a ter valores de features mais homogneos.

De maneira geral os resultados de ambos os mtodos esto proximos, entretando para o caso presente a clusterizao hierquica aglomerativa apresentou uma deliamento de fronteira menos catico no dendograma, deixando claro que os pases so um concorrentes a mudar de categoria, inclusive os outliers aparentam no estarem em seu grupo por acaso, pois por mais pobre que um pas seja ele ainda ser considerado subdesenvolvido, da mesma forma que o mais rico ainda assim estar no grupo de desenvolvidos.

##### Descrio dos algoritmo K-means

Partindo do pressuposto que os dados esto tratados, o algoritmo trabalha da seguinte forma:

1. So escolhidos pontos aleatrios no plano, de acordo com a quantidade de clusters pre-definida, tais pontos so os centros dos clusters;
2. Calcula a distancia de cada ponto para o centro dos clusters;
3. Nomeia o ponto para um cluster de acordo com a distancia dele para o centro do mesmo.
4. Calcula o novo centro para o cluster baseado nos dados recem assimilados.
5. Repita o passo 2 ao 4 at que a variao no deslocamento do centro do cluster esteja abaixo de um limiar previamento estabelecido ou que a quantidade de interraes seja maior do que a pre estabelecida.


Para garantir que o centroid do cluster seja um dado vlido o mtodo de Kmedoide deve ser utilizado, o mesmo funciona com o seguinte algoritmo :
1. Seleciona um numro de dados previamente informado para se tornarem o centroide dos clusters;
2. Calcula a distancia de cada ponto para o centro dos clusters;
3. Nomeia o ponto para um cluster de acordo com a menor distancia dele para o centroide do mesmo;
4. Recalcula as dissimilaridades para verificar qual ponto ser o novo centroide do cluster;
5. Repita o passo 2 ao 4 at que a variao no deslocamento do centro do cluster esteja abaixo de um limiar previamento estabelecido ou que a quantidade de interraes seja maior do que a pre estabelecida.

#### K-means e outliers

Kmeans  um algoritmo de clusterizao sensvel a outliers, pois como mencionado nas clculas anteriores, o calclo do centro de um cluster  baseado na distancia quadratica euclidiana mdia de todos os pontos dentro do mesmo, sendo que todos os pontos dos dados so assimilados por um cluster, logo ao calcular a mdia das distancias a mesma sofrer um desvio por conta dos outliers, com isso o baricentro  deslocado o que pode resultar com iterao seguinte assimilando mais pontos para o cluster.

#### DBSCAN

Algoritmos DBSCAN so mais robustos a outliers porque no necessariamente considera todo os pontos como pertencentes a um cluster. Diferente do Kmeans, o DBSCAN necessita de um nmero minimo de pontos por cluster e um valor de distancia mxima para um pontos ser considerado vizinho do cluster, logo outliers estaro acima desta amplitude e no sero considerados parte de um cluster e com isso o centroide de um cluster no  afetado pelo menos.
